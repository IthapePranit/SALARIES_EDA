### Read Dataset CARS93

## Q7: Find out Models of cars which are having Price between 17 and 25

```python
A[(A.Price>=17)&(A.Price<=25)][["Model"]]
```

## Q8:Find out top 7 most fuel economic cars for highway

```python
A.sort_values(by=['MPG.highway'],ascending=False).head(7)[["Manufacturer","Model","MPG.highway"]]
```

## Q9 :Find out Models of cars having no AirBags and which are of compact type whose price is below 20.

```python
A[(A.AirBags=="None")&(A.Type=="Compact")&(A.Price<20)][["Model"]]
```
